corr() : It is used to compute the pairwise correlation of columns in a DataFrame.

```python
df.corr()
```

.style: Accesses the Styler object which is used to customize the DataFrame's appearance.
.background_gradient(): Applies a color gradient to the background of the DataFrame cells, based on the values.

```python
df.corr().style.background_gradient()
```

The resulting plot is a heatmap where each cell represents the correlation coefficient between pairs of numerical features. 
The colors indicate the strength and direction of the correlations, with annotations showing the exact values.
annot=True: Annotates each cell with the numeric value of the correlation coefficient.
cmap='GnBu': Specifies the color map (Green to Blue) for the heatmap.
fmt='.1f': Formats the annotation text to one decimal place.
linewidths=.6: Sets the width of the lines that divide each cell.

```python
fig, ax = plt.subplots(figsize=(12, 12))
sns.heatmap(df.select_dtypes(include=['int64', 'float64']).corr(), annot=True , cmap='GnBu' , fmt='.1f' , linewidth=.6)
plt.title('Correlation heatmap for numerical features')
plt.show()
```

Selecting the appropriate regression model for the Credit Card Fraud Detection dataset largely depends on the nature of the problem you're trying to solve. Given that this dataset is used for fraud detection, which is a binary classification problem (fraudulent vs. non-fraudulent transactions), you should apply a classification model rather than a regression model.

### Splitting feature and target 

```python
x=df.iloc[:, :-1]
y=df.iloc[:, -1]
```

### Features : All columns except last will be used as features.

```python
x
```

### Target : The last column will be used as the target variable.

```python
y
```

### Logistic Regression

### Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(xtrain,ytrain)
ypred=logreg.predict(xtest)
```

```python
ypred
```

### To calculate Accuracy score, Confusion matrix, Classification report

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

### Calculate accuracy

```python
ac = accuracy_score(ytest, ypred)
```

### Calculate confusion matrix

```python
cm = confusion_matrix(ytest, ypred)
```

### Generate classification report

```python
cr = classification_report(ytest, ypred)
```

```python
print(f"Accuracy: {ac}\n\nConfusion Matrix:\n{cm}\n\nClassification Report:\n{cr}")
```
